```python
import os
import requests
from tqdm import tqdm

# Define the input folder containing .txt files and the base output folder
input_folder = "/content/pose_text_file_left"  # Folder containing all .txt files
output_base_folder = "/content/MAIN_YOGA_DATASET_3"
# Base folder for saving images
os.makedirs(output_base_folder, exist_ok=True)

# Get all .txt files in the input folder
txt_files = [f for f in os.listdir(input_folder) if f.endswith(".txt")]

# Set custom headers (e.g., User-Agent)
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Function to download images from a list of URLs
def download_images(urls, output_folder):
    for url in tqdm(urls, desc="Downloading Images"):
        try:
            image_name = os.path.join(output_folder, url.split("/")[-1])  # Save with original name
            response = requests.get(url, headers=headers, stream=True)  # Include headers here
            if response.status_code == 200:
                with open(image_name, "wb") as img_file:
                    for chunk in response.iter_content(1024):
                        img_file.write(chunk)
            else:
                print(f"Failed to download {url} - Status code: {response.status_code}")
        except Exception as e:
            print(f"Error downloading {url}: {e}")

# Loop through each .txt file in the folder
for txt_file in txt_files:
    file_path = os.path.join(input_folder, txt_file)

    # Create a separate folder for each text file's images
    output_folder = os.path.join(output_base_folder, txt_file.replace(".txt", "_images"))
    os.makedirs(output_folder, exist_ok=True)

    # Read and extract URLs from the current .txt file
    with open(file_path, "r") as file:
        lines = file.read().splitlines()

    # Process lines to extract URLs
    urls = []
    for line in lines:
        parts = line.split("\t")  # Split by tab
        if len(parts) > 1:  # Ensure valid split
            urls.append(parts[1].strip())  # Extract URL, removing any whitespace

    # Download images for the current file
    print(f"Downloading images for {txt_file}...")
    download_images(urls, output_folder)

print("Download complete!")
```
